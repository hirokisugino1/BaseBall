```python
#url情報の取得メソッド
def get_url(url):
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.5249.168 Safari/537.36"}
    re = requests.get(url,headers=headers)
    soup = BeautifulSoup(re.content,'html.parser')
    return soup
base_url = 'https://baseball-data.com'
soup =get_url(base_url)
body = soup.find_all("table")[3]

#取得データの器
pa_piching_players = []

#球団
kyuudan = [i*6 for i in range(6)]

for n in tqdm(kyuudan):
    base_team={}
    #球団毎
    base_team['球団'] = body.find_all("td")[n].getText()
    #選手一覧取得
    players=body.find_all("td")[round(n+3)].find('a')
    url_players = players.get("href")
    #セリーグスープ
    ya_soup=get_url(url_players)

    #選手データ
    for m in range(1,len(ya_soup.find_all("table")[0].find_all('tr'))-2):
        player = base_team.copy()

        #空白文字削除
        text = "a\u3000 b\t\nc\r\n"
        table = str.maketrans({
          '\u3000': '',
          ' ': '',
          '\t': ''
        })
        text = text.translate(table)

        url_ya_soup_s=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[1].find('a').get('href')
        ya_soup_s = get_url(url_ya_soup_s) 

        #背番号
        player['背番号']=ya_soup.find_all("table")[0].find_all("tr")[m].find_all("td")[0].getText()

        #選手名
        player['選手名']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[1].getText().translate(table)

        #防御率
        player['防御率']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[2].getText()

        #試合
        player['試合']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[3].getText()

        #勝利
        player['勝利']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[4].getText()

        #敗北
        player['敗北']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[5].getText()

        #セーブ
        player['セーブ']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[6].getText()

        #ホールド
        player['ホールド']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[7].getText()

        #HP
        player['HP']=ya_soup_s.find_all('table')[1].find_all('tr')[1].find_all('td')[5].getText()

        #完投
        player['完投']=ya_soup_s.find_all('table')[1].find_all('tr')[1].find_all('td')[6].getText()

        #完封勝
        player['完封勝']=ya_soup_s.find_all('table')[1].find_all('tr')[1].find_all('td')[7].getText()

        #無四球
        player['無四球']=ya_soup_s.find_all('table')[1].find_all('tr')[1].find_all('td')[8].getText()

        #敬遠
        player['敬遠']=ya_soup_s.find_all('table')[1].find_all('tr')[1].find_all('td')[15].getText()  
        
        #勝率
        player['勝率']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[8].getText()

        #打者
        player['打者']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[9].getText()

        #投球回
        player['投球回']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[10].getText()

        #被安打
        player['被安打']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[11].getText()

        #被本塁打
        player['被本塁打']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[12].getText()

        #与四球
        player['与四球']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[13].getText()
        
        #与死球
        player['与死球']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[14].getText()
        
        #奪三振
        player['奪三振']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[15].getText()
        
        #暴投
        player['暴投']=ya_soup_s.find_all('table')[1].find_all('tr')[1].find_all('td')[18].getText()

        #ボーク
        player['ボーク']=ya_soup_s.find_all('table')[1].find_all('tr')[1].find_all('td')[19].getText()
        
        #失点
        player['失点']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[16].getText()
        
        #自責点
        player['自責点']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[17].getText()
        
        #WHIP
        player['WHIP']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[18].getText()
        
        #DIPS
        player['DIPS']=ya_soup.find_all("table")[0].find_all('tr')[m].find_all('td')[19].getText()

        #被安打率
        player['被安打率']=ya_soup_s.find_all('table')[2].find_all('tr')[1].find_all('td')[1].getText()

        #被本塁打率
        player['被本塁打率']=ya_soup_s.find_all('table')[2].find_all('tr')[1].find_all('td')[2].getText()

        #奪三振率
        player['奪三振率']=ya_soup_s.find_all('table')[2].find_all('tr')[1].find_all('td')[3].getText()

        #与四球率
        player['与四球率']=ya_soup_s.find_all('table')[2].find_all('tr')[1].find_all('td')[4].getText()

        #K/BB
        player['K/BB']=ya_soup_s.find_all('table')[2].find_all('tr')[1].find_all('td')[6].getText()

        #HidR
        player['HidR']=ya_soup_s.find_all('table')[2].find_all('tr')[1].find_all('td')[7].getText()

        #PFR
        player['PFR']=ya_soup_s.find_all('table')[2].find_all('tr')[1].find_all('td')[8].getText()
        
        pa_piching_players.append(player)
        
time.sleep(1)
```
